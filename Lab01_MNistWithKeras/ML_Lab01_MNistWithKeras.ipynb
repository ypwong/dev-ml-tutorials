## Preprocessing the Image Data - Flatten the Training and Test Images

```python
image_height, image_width = X_train[0].shape
num_of_training_images = y_train.size
num_of_testing_images  = y_test.size

print("image_height =", image_height)
print("image_width  =", image_width)
print("num_of_training_images =", num_of_training_images)
print("num_of_testing_images  =", num_of_testing_images)
```

```python
X_train = X_train.reshape(num_of_training_images, image_height * image_width)
X_test  =  X_test.reshape(num_of_testing_images,  image_height * image_width)

print("Reshaped X_train.shape =", X_train.shape)
print("Reshaped X_test.shape  =", X_test.shape)
```

```python
print(X_train[0].shape)
print(X_train[0])
```

## Preprocessing the Image Data - Normalize the Training and Test Image Data

```python
X_train = X_train.astype("float32")
X_test  = X_test.astype("float32")

X_train /= 255.0
X_test  /= 255.0
print(X_train[0].shape)
print(X_train[0])
```

## Preprocessing the Label Data - Convert Training and Test Labels Into Categorical Vectors

```python
print(y_train.shape)
print(y_test.shape)

# Print the first 3 training and test labels
print(y_train[0], y_train[1], y_train[2])
print(y_test[0],  y_test[1],  y_test[2] )
```

```python
# Convert training and test labels into categorical vectors
y_train = to_categorical(y_train, 10)
y_test  = to_categorical(y_test , 10)

print(y_train.shape)
print(y_test.shape)

# Print the first 3 training and test labels
print(y_train[0], y_train[1], y_train[2])
print(y_test[0],  y_test[1],  y_test[2] )
```

## Build a Model

```python
print("input_shape =", X_train[0].shape)

layer_1_input_shape    = X_train[0].shape
layer_1_num_of_outputs = 512
layer_2_num_of_outputs = 512
layer_3_num_of_outputs = 10
```

```python
model = Sequential()

layer_1 = Dense( layer_1_num_of_outputs, activation="relu", input_shape=layer_1_input_shape )
layer_2 = Dense( layer_2_num_of_outputs, activation="relu")
layer_3 = Dense( layer_3_num_of_outputs, activation="softmax")

model.add(layer_1)
model.add(layer_2)
model.add(layer_3)
```

## Compile the Model

```python
model.compile( optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"] )
```

```python
model.summary()
```

```python
def print_layer_info(layer_no, num_inputs, num_outputs):
   print("Layer Number =>", layer_no)
   print("Number of inputs  =", num_inputs)
   print("Number of outputs =", num_outputs)
   print("Number of parameters =", (num_inputs + 1) * num_outputs)  # + 1 are the biases

print_layer_info( 1, layer_1_input_shape[0], layer_1_num_of_outputs )
print()
print_layer_info( 2, layer_1_num_of_outputs, layer_2_num_of_outputs )
print()
print_layer_info( 3, layer_2_num_of_outputs, layer_3_num_of_outputs )
```

## Train the Model

```python
def train(num_of_epochs, batch_size):
  print("Number of epochs =", num_of_epochs)
  print("Batch size =", batch_size)
  print("Number of batches =", num_of_training_images / batch_size)

  history = model.fit(
                      X_train, y_train,
                      epochs = num_of_epochs,
                      batch_size = batch_size,
                      validation_data = (X_test, y_test)
                     ).history
  return history
```

```python
num_of_epochs = 5

history0 = train(num_of_epochs = num_of_epochs, batch_size = 16)

print(history0)
```

### Accuracy of Training and Validation with Loss

```python
x = list( range(1, num_of_epochs + 1) )
x_range = (1, num_of_epochs + 1)

def plot_accuracies_and_loss(history, batch_size):
  fig, ax1 = plt.subplots()
  ax2 = ax1.twinx()

  ax1.plot(x, history["accuracy"],     color = "#222266", label="Training Accuracy")
  ax1.plot(x, history["val_accuracy"], color = "#2222cc", label="Validaction Accuracy")
  ax2.plot(x, history["loss"],         color = "r",       label="Loss")

  ax1.locator_params(axis='y', nbins=10)
  ax2.locator_params(axis='y', nbins=10)
  ax1.set_xticks(range(1, num_of_epochs + 1, 1))
  ax1.tick_params(axis="x", color="g", labelcolor="g")
  ax1.tick_params(axis="y", color="b", labelcolor="b")
  ax2.tick_params(axis="y", color="r", labelcolor="r")

  ax1.set_xlim(x_range)

  ax1.set_xlabel("Epoch",      color="g")
  ax1.set_ylabel("Accuracies", color="b")
  ax2.set_ylabel("Loss",       color="r")

  fig.suptitle(f"Accuracies & Loss (Batch Size {batch_size})")
  fig.legend(loc='center')

  fig.canvas.draw()
  plt.show()
```
